## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demo = purchase_data.loc[:,["Gender","SN","Age"]]
gender_demo = gender_demo.drop_duplicates()

gender_total = gender_demo["Gender"].value_counts()

gender_percent = gender_total / total_players 

data_frame = pd.DataFrame({"Total Count":gender_total,"Percentage of Players":gender_percent})
table_gender = pd.DataFrame(data_frame)
table_gender
table_gender["Percentage of Players"] = table_gender["Percentage of Players"].map("{:,.2%}".format)
table_gender


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
age_categories = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
gender_demo["Age Range"] = pd.cut(gender_demo["Age"],bins,labels = age_categories)
total_counts = gender_demo["Age Range"].value_counts()
percentage_of_players = total_counts / total_players
age_demo_dataframe = {"Total Count":total_counts,"Percentage of Players":percentage_of_players}
age_demo_dataframe = pd.DataFrame(age_demo_dataframe)
age_demo_dataframe = age_demo_dataframe.sort_index()
age_demo_dataframe
age_demo_dataframe["Percentage of Players"] = age_demo_dataframe["Percentage of Players"].map("{:,.2%}".format)
age_demo_dataframe
#format decimal places
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data framez

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels = age_categories)

total_purchase = purchase_data.groupby(["Age Group"]).sum()["Price"].rename("Total Purchase Value")
avg_price = purchase_data.groupby(["Age Group"]).mean()["Price"].rename("Average Purchase Price")
purchase_count = purchase_data.groupby(["Age Group"]).count()["Price"].rename("Purchase Count")
avg_total_purchase_person = total_purchase / age_demo_dataframe["Total Count"]

age_data = {"Purchase Count":purchase_count,"Average Purchase Price":avg_price,"Total Purchase Value":total_purchase,"Avg Total Purchase per Person":avg_total_purchase_person}
age_data = pd.DataFrame(age_data)
age_data
age_data["Purchase Count"] = age_data["Purchase Count"].map("{:,.2f}".format)
age_data["Average Purchase Price"] = age_data["Average Purchase Price"].map("${:,.2f}".format)
age_data["Total Purchase Value"] = age_data["Total Purchase Value"].map("${:,.2f}".format)
age_data["Avg Total Purchase per Person"] = age_data["Avg Total Purchase per Person"].map("${:,.2f}".format)
age_data
```

## Most Popular Items

* Display a preview of the summary data frame
